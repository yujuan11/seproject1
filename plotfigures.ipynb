```python
import matplotlib.pyplot as plt
```

```python
mcs=data_df[0] # monte carlo step
energy=data_df[2] # reduced energy
para=data_df[3] # order parameter
```

```python
plt.figure(figsize=(8, 6))
plt.plot(mcs,energy,label='Reduced Temperature, T*= 0.65')
plt.xlabel('MCS')
plt.ylabel('Reduced Energy U')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(mcs,para,label='Reduced Temperature, T*= 0.65')
plt.xlabel('MCS')
plt.ylabel('Order Parameter, S')
plt.show()
```
